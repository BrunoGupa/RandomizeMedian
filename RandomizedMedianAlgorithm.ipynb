# Randomized Median Algorithm

```python
import random
import math
# Generate n different random numbers
n = 10000001 #n>30
limit_sup = 10000000
limit_inf = -10000000
S = random.sample(range(limit_inf,limit_sup), n)
S
```

```python
def RandomizedMedian(S):
    n = len(S)
    special_number = n**(3/4)
    ceil_sn = math.ceil(special_number)
    # Pick an indepedent and uniformly sample with replacement
    R = random.choices(S,k=ceil_sn)
    R.sort()
    d = math.floor(1/2*special_number - math.sqrt(n))
    u = math.ceil(1/2*special_number + math.sqrt(n))
    d,u = R[d],R[u]
    C = []
    l_d = 0
    l_u = 0
    for i in S:
        if i < d:
            l_d += 1
        elif i > u:
            l_u += 1
        else:
            C.append(i)
    if (l_d > n/2) or (l_u > n/2):
        print('FAIL')
    if len(C) <= 4 * special_number:
        C.sort()
    else:
        print('FAIL')
    median = (math.floor(n/2) - l_d + 1) - 1
    return C[median]
```

```python
# Brut Force algorithm
def Median(S):
    S.sort()
    n = len(S)
    p = math.floor(n/2)
    if n % 2 == 0:
        return S[p-1]
    else:
        return S[p]
```
